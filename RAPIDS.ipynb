```python
X = df.drop(["CATEGORY_ID"],axis=1)
X = X[:7000]
y = df["CATEGORY_ID"][:7000]

X_cudf, X_cudf_test, y_cudf, y_cudf_test = train_test_split(X, y, test_size = 0.3, random_state=0)

print(type(X_cudf))
print(type(y_cudf))
print(type(X_cudf_test))
print(type(y_cudf_test))

print(X_cudf.shape)
print(y_cudf.shape)
print(X_cudf_test.shape)
print(y_cudf_test.shape)
```

```python
tfidf_model = TfidfVectorizer(analyzer = "char", ngram_range = (1,2),max_df = 0.9975,
                              min_df = 0.0025,
                              binary = True,
                              norm = "l1",
                              use_idf =  True,
                              smooth_idf = True,
                              sublinear_tf = True)

X_cudf_vec = tfidf_model.fit_transform(X_cudf.lemma).toarray()
X_cudf_vec_test = tfidf_model.transform(X_cudf_test.lemma).toarray()


print('train text embeddings shape is{}, \n test text embeddings shape is: {}'.format(X_cudf_vec.shape, X_cudf_vec_test.shape) )
print("y_cudf and y_cudf_test shape:{} {}".format(y_cudf.shape, y_cudf_test.shape))
```

```python
model = SVC(gamma="auto", multiclass_strategy="ovr", kernel="rbf")

model.fit(X_cudf_vec, y_cudf)
```
